For basic plots like line or bar charts, `matplotlib.pyplot` is imported as `plt`. `seaborn`, imported as `sns`, is used for advanced plots.

## Line Charts
Simplest form of data visualization. Data points connected using straight lines. Shape, size, color, and other elements are customizable. 

Let's take a simple example of imaginary data for yields of apples (ton per hectare) over 6 years.

```python
apple_yields = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931]
```

`plt.plot` function plots a line start. If one argument is given, it is taken along y axis. If 2 arguments are given, first one is for x axis and the other one is for y axis.

```python
plt.plot(apple_yields)
```

`[<matplotlib.lines.Line2D at 0x12ac45880>]` is the object type that is returned by `plt.plot`. Adding a semicolon(;) to the end of statement removes this.

### Customizing the X-axis
In the above plot, the x axis is populated with indices of the `apple-yields` list. Lets add years instead.

```python
years = [2010, 2011, 2012, 2013, 2014, 2015]
plt.plot(years, apple_yields);
```

The yields have gradually increased till 2013. After that it looks like it will be reach a saturation point at around 0.95. This might be due to usage of old techniques, weather, etc.

### Axis Labels

Labels or names can be added to the graph using `plt.xlabel` and `plt.ylabel` functions. In Jupyter Notebook, a single cell with all plt.* functions will bbe applied to the same graph.

```python
plt.plot(years, apple_yields)
plt.xlabel('Years')
plt.ylabel('Yield (tons per hectare)');
```

### Plotting Multiple Lines
This is simple. Just invoke `plt.plot` function multiple times.

```python
years = range(2000, 2012)
apples = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931, 0.934, 0.936, 0.937, 0.9375, 0.9372, 0.939]
oranges = [0.962, 0.941, 0.930, 0.923, 0.918, 0.908, 0.907, 0.904, 0.901, 0.898, 0.9, 0.896, ]
```

```python
plt.plot(years, apples)
plt.plot(years, oranges)
plt.xlabel('Years')
plt.ylabel('Yield (tons per hectare)');
```

This shows that the yield for oranges is decreasing while the yield for apples is increasing. This might be due to various reasons. One of them can be that the demand for oranges is decreasing and for apples it is increasing. As a result, the land for orange farming is now used for apples. These is a hypothesis. Reality check can be done by further investigating the scenario.

### Chart Title and Legend
In the above chart, it is not clear which line is for apple and which is for orange. To make is easier to understand, legends are added. `plt.legend` is used for the same. `plt.title` can make the graph more informative by adding a title.

```python
plt.plot(years, apples)
plt.plot(years, oranges)

plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')

plt.title("Crop Yields in Kanto")
plt.legend(['Apples', 'Oranges']);
```

`plt.legend` takes a list as argument. The first element should be the legend for the first plot, and so on.

### Line Markers
What was the yield for apples in the year 2015? Well, making an estimation using the above graph is slightly tedious. Let's make it more informative by adding markers.

`plt.plot` uses `marker` argument to mark the data points. They can be of many styles like circle, cross, square, diamond, etc. The list of supported markers is available here: https://matplotlib.org/3.1.1/api/markers_api.html

```python
plt.plot(years, apples, marker='o')
plt.plot(years, oranges, marker='x')

plt.xlabel('Years')
plt.ylabel('Yield (tons per hectare)')

plt.title('Crop Yields in Kanto')
plt.legend(['Apples','Oranges']);
```

### Styling lines and markers
`plt.plot` supports many arguments for styling lines and markers.
* `color` or `c`: set the color for line ([supported colors](https://matplotlib.org/3.1.0/gallery/color/named_colors.html))
* `linestyle` or `ls`: choose between a solid or dashed line
* `linewidth` or `lw`: set the width of line
* `markersize` or `ms`: set the size of markers
* `markeredgecolor` or `mec`: set the edge color for markers
* `markeredgewidth` or `mew`: set the edge width for markers
* `markerfacecolor` or `mfc`: set the fill color for markers
* `alpha`: opacity of the plot

Check out the documentation for plt.plot to learn more: https://matplotlib.org/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib.pyplot.plot

```python
plt.plot(years, apples, marker='s', c='b', ls='-', lw=2, ms=8, mew=2, mec='navy')
plt.plot(years, oranges, marker='o', c='r', ls='--', lw=3, ms=10, alpha=.5)

plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')

plt.title("Crop Yields in Kanto")
plt.legend(['Apples', 'Oranges']);
```

The `fmt` provides a shorthand for specifying the marker, line style and color. It is the third argument in `plt.plot`. Hence, there is no need to mention the paramter keyword. Syntax for `fmt` is: `[marker][line][color]`

```python
# Storing in function to make it easier to demonstrate other features of plot
def fruits_plot():
    plt.plot(years, apples, 's-b') # square markers, unbroken line and blue
    plt.plot(years, oranges, 'o--r') # round markers, dashed line and red

    plt.xlabel('Year')
    plt.ylabel('Yield (tons per hectare)')

    plt.title("Crop Yields in Kanto")
    plt.legend(['Apples', 'Oranges']);
fruits_plot()
```

If no line style is specified, then `fmt` does not draw a line. The same goes for all other parts of the parameter. Giving an empty string, is as good as leaving out the fmt argument. 

```python
plt.plot(years, oranges, 'or');
```

### Changing the figure size
`plt.figure` can change the plot size.

```python
plt.figure(figsize=(12,6))
plt.plot(years, oranges)
plt.title('Yield of Oranges (tons per hectare)');
```

### Improving default styles using Seaborn
Using Seaborn is an easy way to change styles globally to make the charts look beautiful. A full list of predefined styles is here: https://seaborn.pydata.org/generated/seaborn.set_style.html. A style from this list can be set using `sns.set_style`.

```python
# Example 1
sns.set_style('whitegrid')
fruits_plot()
```

```python
# Example 2
sns.set_style('darkgrid')
fruits_plot()
```

These styles make the estimation for in between point much easier. Matplotlib also provides option to change the paramters for chart display globally, using `matplotlib.rcParams`. Learn more: https://matplotlib.org/3.2.1/tutorials/introductory/customizing.html#matplotlib-rcparams

```python
import matplotlib
matplotlib.rcParams
```

```python
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (9,5)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

```python
fruits_plot()
```

This plot is a fusion of Seaborn and Matplotlib display settings.
